  - Descubra cuántas quejas se recibieron y enviaron el mismo día

2. Get the complaints received in the states of new york
   - Obtenga las quejas recibidas en los estados de nueva york

3. Get the complaints received in the states of new york and California
  - Obtenga las quejas recibidas en los estados de nueva york y California

4. Get all rows with the word "Credit" in the Product Field
  - Obtenga todas las filas con la palabra "Crédito" en el campo Producto

5. Get all rows with the word "Late" in the Issue Field
  - Obtenga todas las filas con la palabra "Crédito" en el campo de emisión

```
# File location and type
file_location = "/FileStore/tables/ConsumerComplaints-1.csv"
file_type = "csv"

# CSV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
df = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(df)
```

```
from pyspark.sql.functions import count, desc , col, max, struct
```

```
df.printSchema()
```

```
# Find out how many complaints were received and sent on the same day

q1 = df.select('Date_Received', 'Date_Sent_to_Company').filter(df.Date_Received == df.Date_Sent_to_Company).groupby('Date_Received').agg(count('Date_Received')).alias('count')

display(q1)
```

```
#Get the complaints received in the states of new york

q2 = df.select('State_Name').filter(df.State_Name == 'NY').groupby('State_Name').agg(count('State_Name')).alias('count')

display(q2)
```

```
#Get the complaints received in the states of new york and California

q3 = df.select('State_Name').filter((df.State_Name == 'NY') | (df.State_Name == 'CA')).groupby('State_Name').agg(count('State_Name')).alias('count')
display(q3)
```

```
#Get all rows with the word "Credit" in the Product Field

q4 = df.select('*').filter(df.Product_Name.like('%Credit%'))

display(q4)
```

```
#Get all rows with the word "Late" in the Issue Field
q5 = df.select('*').filter(df.Issue.like('%Late%'))

display(q5)
```
